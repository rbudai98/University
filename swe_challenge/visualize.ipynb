```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

x_dim = 50
y_dim = 50
```

```python
def load_np(name):
	B_load = np.load(name)
	B = np.zeros((x_dim, y_dim, 3))

	for i in range(x_dim):
		for j in range(y_dim):
			for k in range(3):
				B[i,j,k] = B_load[(i*x_dim + j)*3 + k]
	return B
```

```python
# Load data
B = load_np('processed_data/B_Lorentzian.npy')
B_rough = np.load('processed_data/B_rough.npy')

B_mag = load_np('processed_data/B_mag_Lorentzian.npy')
B_mag_rough = np.load('processed_data/B_mag_rough.npy')
```

```python
import plotly.graph_objs as go
import numpy as np 

B_x = B[:,:,0]
B_y = B[:,:,1]
B_z = B[:,:,2]

print(B_x)

x_points = []
y_points = []
z_points = []

for i in range(x_dim):
	for j in range(y_dim):
		x_points.append(np.array(i+0, i+B_x[i][j]))
		y_points.append(np.array(j+0, j+B_y[i][j]))
		z_points.append(np.array(0, B_z[i][j]))

fig = go.Figure(data=go.Scatter3d(x=x_points, y=y_points,z=z_points, mode='lines'))
fig.show() 
```
